## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.wait import WebDriverWait
import time
browser = webdriver.Chrome(executable_path="C:/Users/Username/Desktop/program/python/1st-PyCrawlerMarathon/chromedriver")
browser.get("https://www.ettoday.net/news/news-list.htm")
browser.implicitly_wait(10)
today=time.strftime("%Y/%m/%d", time.localtime())#https://www.runoob.com/python/python-date-time.html
while True:
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if today not in soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text:
        break
    time.sleep(2)
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if today in d.find(class_="date").text:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.wait import WebDriverWait
browser = webdriver.Chrome(executable_path="C:/Users/Username/Desktop/program/python/1st-PyCrawlerMarathon/chromedriver")
browser.get("https://www.ettoday.net/news/news-list.htm")
browser.implicitly_wait(10)
now=time.time()

while True:
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    news_time=soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text+':00'
    news_time = time.strptime(news_time, "%Y/%m/%d %H:%M:%S")
    news_time=time.mktime(news_time)#https://blog.csdn.net/holdlg/article/details/62436537 轉成時間戳
    if now-news_time>2*60*60:
        break
    time.sleep(2)
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time=d.find(class_="date").text+':00'
    news_time=time.strptime(news_time, "%Y/%m/%d %H:%M:%S")
    news_time=time.mktime(news_time)
    if now-news_time<2*60*60:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import datetime
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.wait import WebDriverWait
now = datetime.datetime.now()
delta = datetime.timedelta(days=3)
news_time=now-delta
three_oclock=news_time.strftime('%Y/%m/%d')+' 15:00:00'
three_oclock=time.strptime(three_oclock, "%Y/%m/%d %H:%M:%S")
three_oclock=time.mktime(three_oclock)
five_oclock=news_time.strftime('%Y/%m/%d')+' 17:00:00'
five_oclock=time.strptime(five_oclock, "%Y/%m/%d %H:%M:%S")
five_oclock=time.mktime(five_oclock)

browser = webdriver.Chrome(executable_path="C:/Users/Username/Desktop/program/python/1st-PyCrawlerMarathon/chromedriver")
url='https://www.ettoday.net/news/news-list-'+news_time.strftime('%Y-%m-%d')+'-0.htm'
browser.get(url)
browser.implicitly_wait(10)

while True:
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    news_time=soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text+':00'
    news_time = time.strptime(news_time, "%Y/%m/%d %H:%M:%S")
    news_time=time.mktime(news_time)#https://blog.csdn.net/holdlg/article/details/62436537
    if news_time<three_oclock:
        break
    time.sleep(2)
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time=d.find(class_="date").text+':00'
    news_time=time.strptime(news_time, "%Y/%m/%d %H:%M:%S")
    news_time=time.mktime(news_time)
    if three_oclock<news_time<five_oclock:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

browser.quit()
```
