```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data=json.loads(r.text)
print(data)
i=0
for d in data:
    i+=1
print(i)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for d in data:
    print(d['title'],d['createdAt'],d['commentCount'],d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
avecomment=[]
avelike=[]
for d in data:
    avecomment.append(d['commentCount'])
    avelike.append(d['likeCount'])
print('å¹³å‡ç•™è¨€äººæ•¸',sum(avecomment)/len(avecomment),'\n','å¹³å‡æŒ‰è®šäººæ•¸',sum(avelike)/len(avelike))
```
